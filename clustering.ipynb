# $K-Means$聚类分析

1. 聚类是什么：把样本划分为由相似的对象组成的多个类的过程
2. 聚类算法能做什么：
3. 聚类和分类之间的区别：分类问题：比如支持向量机对iris数据集以及乳腺癌数据集的分类，已知类别，目标是找出分类超平面，来预测测试数据的类别。而在聚类问题之中，数据的类别未知。需要将数据分成不同的类别，以便更准取地使用统计模型进行估计、分析以及预测。或者是分析不同类别之间的相关性与差异。
4. `make_clssification()`：生成n_samples个数据，每个数据具有n_features个特征。同时，可以指定有效特征，冗余特征，重复特征的数量。指定数据的类别，每类数据的聚类中心

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
from sklearn.datasets import make_blobs

x_train, y_train = make_blobs(n_samples=300, centers=4, cluster_std=1, random_state=1)
plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train, s=10)
```

```python
model = KMeans(n_clusters=4)
model.fit(x_train)
```

```python
y_hat = model.predict(x_train)
y_hat
```

```python
def show_acc(y_actual, y_hat):
    acc = y_actual == y_hat
    return np.mean(acc)
```

```python
from numpy import unique
from numpy import where
from sklearn.datasets import make_classification
from sklearn.cluster import MiniBatchKMeans
from matplotlib import pyplot
# 定义数据集
X, _ = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=4)
# 定义模型
model = MiniBatchKMeans(n_clusters=2)
# 模型拟合
model.fit(X)
# 为每个示例分配一个集群
yhat = model.predict(X)
# 检索唯一群集
clusters = unique(yhat)
# 为每个群集的样本创建散点图
for cluster in clusters:
# 获取此群集的示例的行索引
    row_ix = where(yhat == cluster)
# 创建这些样本的散布
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
# 绘制散点图
pyplot.show()
```
